## Task 1

- Design a similarity measure that calculates similarity between any two digit images. 
- <font color='blue'> **Code** </font> your similarity measure as a function "sim(x1, x2)" where x1 and x2 are two images and your function should return the similarity value. 
- <font color='blue'> **Explain:** </font> give a short explanation on your design. 

<p><font color='red'> You must design your own similarity and CANNOT use any of the following standard measures: </font></p>

- sim(x1, x2) = $x1\circ x2$
- sim(x1, x2) = $e^{(-||x1-x2||^2/L)}$
- sim(x1, x2) = $(x1\circ x2 + 1)^p$

where $\circ$ means vector dot product.

```python
def bits_on_count(x):
    return bin(x).count('1')

def sim(x1, x2):
    # define similarity as the bit count where the same bit is on in both images
    # the lower the value, the more similiar is the two images being compared.
    score = 0
    for i in range(x1.shape[0]):
        score += bits_on_count(int(floor(x1[i]))^int(floor(x2[i])))
    
    return score
```

```python
scores = list()
k_folds = cross_validation.KFold(X.shape[0], n_folds=10)

for train_index, test_index in k_folds:
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]
    
    #compute similarity matrix as kernel
    n = len(train_index)
    print n
    similarity_matrix_train = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            similarity_matrix_train[i,j] = sim(X_train[i], X_train[j])

    m = len(test_index)
    similarity_matrix_test = np.zeros((n, n))
    for i in range(n):
        for j in range(m):
            similarity_matrix_test[i,j] = sim(Y_train[i], X_train[j])

    # fit model
    svc = SVC(kernel='precomputed')
    fit = svc.fit(similarity_matrix_train, Y_train)
    
    # determine accuracy
    score = fit.score(similarity_matrix_test)
    scores.append(score)
    #print score
    
print "10-folds cross-validation accuracy = %0.3f" % np.mean(scores)
```

```python
digits_4_6_indices = np.where((Y==4) | (Y==6))
images_4_6 = X[digits_4_6_indices]
labels_4_6 = Y[digits_4_6_indices]
print images_4_6.shape
print images_4_6
```

```python
# compute similarity matrix
n = images_4_6.shape[0]
similarity_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        similarity_matrix[i,j] = sim(images_4_6[i], images_4_6[j])
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# use the same threshold as before to determine matching digit 
similarity_threshold = 70
for i in range(0,n):
    # get all images that are supposed to be the same as image of interest
    matches = labels_4_6[np.where(similarity_matrix[1,:] < similarity_threshold)]
    print "Digit ", labels_4_6[i], "has ", len(matches), " matches."
        
    #create an empty graph
    G = nx.Graph()
    prev_node = "%d.%03d"%(matches[0], 0)
    G.add_node(prev_node)
    for j in range(0, len(matches)):
        current_node = "%d.%03d"%(matches[j], j)
        G.add_node(current_node)
        if (matches[j] == matches[j-1]):
            G.add_edge(prev_node, current_node)
        prev_node = current_node
    
    print("graph has %d nodes with %d edges, %d connected components."\
         %(nx.number_of_nodes(G),nx.number_of_edges(G),nx.number_connected_components(G)))
    
    for cc in nx.connected_component_subgraphs(G):
        nodes = {}
        for c in cc.nodes():
            print nodes.append(c)
        
        print "component has %d nodes: members={%s}"%(len(cc.nodes()), nodes.join(,))

    
    pos = nx.spring_layout(G, k=0.315,iterations=200) #nx.random_layout(G) #
    nx.draw(G, pos, node_size=800, with_labels=True, font_size=8)
    plt.show()
    
```
